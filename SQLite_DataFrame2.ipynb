```python
import pandas as pd
import sqlite3 as sql
```

```python
!wget "https://github.com/anvarnarz/praktikum_datasets/raw/main/electricity.db"
```

```python
db = sql.connect("electricity.db")
cursor = db.cursor()
```

```python
cursor.execute("SELECT  name  FROM sqlite_master WHERE type='table'")
print(cursor.fetchall())
```

```python
cursor.execute('PRAGMA table_info(elect);')
cursor.fetchall()
```

```python
pd.read_sql("SELECT * FROM elect LIMIT 15",db,index_col = "index")
```

```python
query = "SELECT DISTINCT country FROM elect"
countries = pd.read_sql(query,db)
```

```python
'Uzbekistan' in countries['country'].values
```

```python
query  = "SELECT country,year,coal,gas,hydro,oil FROM elect WHERE country = 'Uzbekistan' AND year>2010"
dfuz = pd.read_sql(query,db)
dfuz
```

**Between**

```python
query  = "SELECT country,year,coal,gas,hydro,oil FROM elect WHERE country = 'Uzbekistan' AND year BETWEEN 2000 AND 2010"
dfuz = pd.read_sql(query,db)
dfuz
```

```python
query  = "SELECT country,year,coal,gas,hydro,oil FROM elect WHERE country = 'Uzbekistan' AND year IN(1990,1995,1996,1999)"
dfuz = pd.read_sql(query,db)
dfuz
```

**lIKE**


```python
query  = "SELECT country,year,coal,gas,hydro,oil FROM elect WHERE country LIKE '%stan' AND year IN(1990,1995,1996,1999) "
dfuz = pd.read_sql(query,db)
dfuz
```
